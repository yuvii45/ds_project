```python
import joblib
from sklearn.preprocessing import StandardScaler

scaler = joblib.load("knn_13_scaler.joblib")
model = joblib.load("knn_13_scaled.joblib")
```

```python
columns = list(range(13))
```

```python
import os
import pandas as pd

# Initialize an empty list to store DataFrames
data_list = []
file_names = []

for file in os.listdir("MFCC-files-v2"):
    file_name = os.path.join("MFCC-files-v2", file)
    data = pd.read_csv(file_name, header=None).T

    # Use only the first 13 columns
    data = data.iloc[:, :13]
    data.columns = columns  # Ensure columns match

    data_list.append(data)

    # Track file name with each row for later result mapping
    file_names.extend([file] * len(data))

# Concatenate all DataFrames
test_df = pd.concat(data_list, ignore_index=True)

# Scale the data
scaled_test_df = pd.DataFrame(scaler.transform(test_df), columns=columns)

# Predict probabilities for all data at once
predict_proba = model.predict_proba(scaled_test_df)

# Prepare the results DataFrame
results = []
for i, proba in enumerate(predict_proba):
    results.append({"file": file_names[i], **{f"prob_class_{j}": proba[j] for j in range(7)}})

results_df = pd.DataFrame(results)

# Display or save the results
print(results_df.head())
```

```python
grouped = results_df.groupby('file').mean().reset_index()
grouped['file'] = (grouped['file'].str[:-9]).astype(int)
grouped.sort_values(by='file', axis=0, inplace=True)
grouped.head()
```

```python
train_label_0 = train[train['label'] == 0.0]

# Step 3: Duplicate the filtered rows
df_duplicates = train_label_0.copy()

# Step 4: Append the duplicated rows back to the original DataFrame
df_augmented = pd.concat([train, df_duplicates], ignore_index=True)

```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler


# Select only the first 13 columns for features
X = df_augmented[[str(i) for i in range(13)]]
y = df_augmented['label']
X = X.to_numpy()

scaler = StandardScaler()
X_standardized = scaler.fit_transform(X)

# Train the model
model = RandomForestClassifier(random_state=42, verbose=2)
model.fit(X_standardized, y)


```
