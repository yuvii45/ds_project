```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Assuming you have a DataFrame `df` with the probabilities for each class
# For example, df = pd.DataFrame(probabilities, columns=['class_1', 'class_2', 'class_3', ..., 'class_7'])
df = pd.read_csv("final_results_knn.csv").drop(['file', 'Class', "Prob", 'Unnamed: 0'], axis=1)

df.columns = [
    "National Anthem",
    "Bhav Geet",
    "Lavani Songs",
    "Asha Bhosale",
    "Kishor Kumar",
    "Michael Jackson",
    "Asha and Kishor"
]

plt.figure(figsize=(13, 10))
# Visualize pairwise correlation
sns.heatmap(df.corr(), annot=True, cmap='Blues', linewidths=0.5)
plt.title("Correlation Heatmap of Class Probabilities")
plt.show()

# Pairwise scatterplot for probability distributions
# sns.pairplot(df)
# plt.show()
```
